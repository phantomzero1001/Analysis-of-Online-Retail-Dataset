```python
#Loading neccesary packages
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import *
```

### Note that I have used preprocessed online data, this preprocessing task I have done in Excel 

```python

path = r"C:\Users\SHRIDHAR\Downloads\Project 2021\DATA\Online Retail_1.xlsx"
data = pd.read_excel(path,sheet_name='Preprocessed Online Retail')
data.head()
```

```python
data.shape
```

```python
# Encoding data
DF = (data.groupby(['InvoiceNo', 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
DF.head()
```

```python
#converting all positive vaues to 1 and everything else to 0
def my_encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
DF_sets = DF.applymap(my_encode_units)
```

```python
# Building the model
frequent_itemsets = fpgrowth(DF_sets, min_support=0.001, use_colnames=True,verbose=0)
frequent_itemsets
```

```python
frequent_itemsets_rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.05)
frequent_itemsets_rules
```

```python
def Get_recommendations(product):
   
    Rules = frequent_itemsets_rules[frequent_itemsets_rules['antecedents'] == frozenset({product})].sort_values(by=['confidence'],ascending=False).head(5)
    
  
    return Rules
    
    
```

```python
Get_recommendations('6 RIBBONS RUSTIC CHARM')
```
